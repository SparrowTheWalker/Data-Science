```python
import pandas as pd
df = pd.read_csv('Pivot Table Tutorial.csv')
df.head()
```

```python
df.pivot_table(index='FirstName',columns='ProductType',values='SubTotal',aggfunc='sum')
```

```python
df.pivot_table(index='FirstName',columns='ProductType',values=['Quantity','SubTotal'],aggfunc='sum')
```
